```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type",axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

### Dropout ve EarlyStopping Overfitting problemleri Ã§Ã¶zmek iÃ§in kullanÄ±lÄ±r.

```python
x_train.shape
```

#### KaÃ§ kolona sahipsek layerÄ±n o deÄŸer de olmasÄ± Ã¶nerilir. (Bu durum deÄŸiÅŸebilir fakat ilk girdi kolon sayÄ±sÄ± kadar olursa iyi olur) units=30

#### Son units=1 olmalÄ± arasÄ± da ilk units deÄŸeri ve son units deÄŸeri arasÄ±nda olsa iyi olur.

#### Sigmoid 0-1 arasÄ± deÄŸer verir. 

#### 0 ve 1 deÄŸer iÅŸlemi yani binary iÅŸlemi yapÄ±ldÄ±ÄŸÄ±ndan loss="binary_crossentropy"

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

#### overfitting yapÄ±ldÄ±ÄŸÄ±ndan 700 bÃ¼yÃ¼k deÄŸer.

```python
model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1)
```

#### Bu durumda 0 ve 1'e ulaÅŸmak yerine elindeki verilere uydurmaya Ã§alÄ±ÅŸÄ±r.Yeni veri verildiÄŸinde yanlÄ±ÅŸ sonuÃ§ verir.

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

#### EÄŸer val_loss artar dÃ¼zeni bozulursa erken durdurur ---> earlystopping

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

```python
tahminlerimiz=model.predict(x_test)
```

```python
tahminlerimiz = np.round(tahminlerimiz)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,tahminlerimiz))
```

#### precision ---> %90 & %92 oranÄ±nda doÄŸru tahmin ediyor.

```python
print(confusion_matrix(y_test,tahminlerimiz))
```
