```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type",axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

### Dropout ve EarlyStopping Overfitting problemleri çözmek için kullanılır.

```python
x_train.shape
```

#### Kaç kolona sahipsek layerın o değer de olması önerilir. (Bu durum değişebilir fakat ilk girdi kolon sayısı kadar olursa iyi olur) units=30

#### Son units=1 olmalı arası da ilk units değeri ve son units değeri arasında olsa iyi olur.

#### Sigmoid 0-1 arası değer verir. 

#### 0 ve 1 değer işlemi yani binary işlemi yapıldığından loss="binary_crossentropy"

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

#### overfitting yapıldığından 700 büyük değer.

```python
model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
model.history.history
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

#### Bu durumda 0 ve 1'e ulaşmak yerine elindeki verilere uydurmaya çalışır.Yeni veri verildiğinde yanlış sonuç verir.

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

#### Eğer val_loss artar düzeni bozulursa erken durdurur ---> earlystopping

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot()
```

```python
tahminlerimiz=model.predict(x_test)
```

```python
tahminlerimiz = np.round(tahminlerimiz)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,tahminlerimiz))
```

#### precision ---> %90 & %92 oranında doğru tahmin ediyor.

```python
print(confusion_matrix(y_test,tahminlerimiz))
```
