```python
# np.pi = pi sayısı
x = np.linspace(-np.pi,np.pi,50)
```

```python
x
```

```python
# İkisini yazarakgrafikleri üst üste çizdirebiliriz.
plt.plot(x,np.sin(x),label='cosinüs')
plt.plot(x,np.cos(x),label='sinüs')
plt.legend()
```

```python
plt.plot(x,np.sin(x),
         label='cosinüs', 
         linestyle = '--', 
         color='red', 
         linewidth = 3, 
         alpha=0.2)

plt.plot(x,np.cos(x),
         label='sinüs', 
         linestyle = '-.', 
         color='blue')
plt.legend()
```

## style_sheets (matplotlib.org)

```python
plt.plot(x,np.sin(x),
         label='cosinüs', 
         linestyle = '--', 
         color='red', 
         linewidth = 3, 
         alpha=0.2)

plt.plot(x,np.cos(x),
         label='sinüs', 
         linestyle = '-.', 
         color='blue')
plt.legend()
plt.xlim(-2,2) # x eksenine limit verilir.
plt.ylim(-0.5,1)
plt.axis('tight') # Sınırlama fonksiyonu
plt.xlabel('x')
plt.ylabel('cos(x)-sin(x)')
plt.title('Cosinüs ve sinüs grafiği')
```

```python
np.random.seed(12345)
x1_koordinatlari = np.random.randint(10,size = 20)
y1_koordinatlari = np.random.randint(5,size = 20)

x2_koordinatlari = np.random.randint(10,size = 20)
y2_koordinatlari = np.random.randint(5, size = 20)

```

```python
x1_koordinatlari
```

```python
y1_koordinatlari
```

```python
# Kesişim noktalarına çubuk.
plt.bar(x1_koordinatlari, y1_koordinatlari, label='1.koordinatlar')
plt.bar(x2_koordinatlari,y2_koordinatlari, label='1.koordinatlar')
plt.legend(loc=2)
```

```python
x = np.arange(3)
plt.bar(x,height=[1,2,3])
plt.xticks(x,['a','b','c'])
```

```python
np.random.seed(12345)
kadin_yaslar = np.random.randint(low=1, high=75, size=100)
erkek_yaslar = np.random.randint(low=1, high=75, size=100)
```

```python
plt.hist(kadin_yaslar,label='kadın yaşları')
plt.hist(erkek_yaslar,label='erkek yaşları')
plt.legend() # Açıklama ekler.
```

```python
np.random.seed(12345)
x = np.random.normal(loc=0, scale=1, size=10000)
plt.hist(x, density=True, bins=30); # density --> yoğunluk
```

```python
fig, ax = plt.subplots(2,2,sharex=True,sharey=True)
x = np.arange(10)
y1 = np.random.rand(10)
y2 = np.random.rand(10)
ax[0,0].plot(x,y1,color='r')
ax[0,1].scatter(x,y2)
ax[1,0].hist(y1)
ax[1,1].barh(x,y2) # Yatay eksende çubuklar.
```
