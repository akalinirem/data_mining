```python
dataFrame = dataFrame.drop("transmission",axis=1) # transmisson sütununu düşür.
```

```python
y = dataFrame["price"].values
x = dataFrame.drop("price",axis=1).values # price sütununu düşür diğer hepsini al.
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
dataFrame.head()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1)) # Çıkış katmanı.

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train, y = y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300) # batch_size ---> verileri part part vermek.
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

#### Yukarıdan aşağıya doğru iniyor mu? 0'a yaklaşıyor mu?

#### los(kayıp) ve val_loss(doğrulama kaybı) birlikte hareket ediyor mu? (Özellikle sonlara doğru.)

#### epochs çok verilirse(Ör,600,1000) overfitting(aşırı öğrenme) durumu oluşabilir.(Nöronlar asıl problemi çözmek yerine en uygun ağırlığı bulmaya çalışıyor bu da tahmin etmemesine neden oluyor.)

#### Eğer belli bir yerden sonra loss ve val_loss çizgileri ayrılıyorsa farklı işlemler yapılması gerekir.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```
